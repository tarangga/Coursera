```python
from scipy import stats
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('scores.csv')
df
```

```python
fm = df.query('gender=="female"')['math score']
m = df.query('gender=="male"')['math score']
fm.size, m.size
```

```python
def compute_tp(arr1, arr2):
  t, p = 0, 0
  mean1, mean2 = np.mean(arr1), np.mean(arr2)
  s1, s2 = np.std(arr1), np.std(arr2)

  var = (((arr1.size-1)*(s1**2) + (arr2.size-1)*(s2**2)))/(arr1.size + arr2.size-2)
  sp = np.sqrt(var)

  ste = sp * np.sqrt(1/arr1.size + 1/arr2.size)

  t = (mean1-mean2)/ste
  df = arr1.size + arr2.size - 2

  p = 2*(1 - stats.t.cdf(t, df))
  return t, p
compute_tp(m, fm)
```

```python
def remove_outlier(arr):
  q3, q1 = np.percentile(arr, [75, 25])
  iqr = q3 - q1
  low = q1 - 1.5*iqr
  up = q3 + 1.5*iqr
  return arr[(arr>=low) & (arr<=up)]
```

```python
m = remove_outlier(m)
fm = remove_outlier(fm)

compute_tp(m, fm)
```
